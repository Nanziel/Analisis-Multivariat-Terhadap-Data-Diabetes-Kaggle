## **Menampilkan Persentase Pria dan Wanita yang tidak diabetes, prediabetes, dan diabetes**

```python
import pandas as pd

# Memastikan Kolom yang dibutuhkan ada di Dataset
if 'Sex' in df.columns and 'Diabetes_012' in df.columns:

    # Memisahkan Data Berdasarkan Jenis Kelamin
    women = df[df['Sex'] == 0]  # 0 kode untuk wanita
    men = df[df['Sex'] == 1]    # 1 kode untuk pria

    # Menghitung Persentase Kategori Diabetes_012 untuk Wanita
    women_counts = women['Diabetes_012'].value_counts(normalize=True) * 100
    rate_women_no_diabetes = women_counts.get(0, 0) # Mengambil persentase wanita yang tidak diabetes (kode = 0)
    rate_women_prediabetes = women_counts.get(1, 0) # Mengambil persentase wanita yang prediabetes (kode = 1)
    rate_women_diabetes = women_counts.get(2, 0)    # Mengambil persentase wanita yang diabetes (kode = 2)

    # Menghitung Persentase Kategori Diabetes_012 untuk Pria
    men_counts = men['Diabetes_012'].value_counts(normalize=True) * 100
    rate_men_no_diabetes = men_counts.get(0, 0)  # Mengambil persentase wanita yang tidak diabetes (kode = 0)
    rate_men_prediabetes = men_counts.get(1, 0)  # Mengambil persentase wanita yang prediabetes (kode = 1)
    rate_men_diabetes = men_counts.get(2, 0)     # Mengambil persentase wanita yang diabetes (kode = 2)

    # Menampilkan hasil
    print(f"Persentase wanita yang tidak diabetes : {rate_women_no_diabetes:.2f}%")
    print(f"Persentase wanita yang prediabetes : {rate_women_prediabetes:.2f}%")
    print(f"Persentase wanita yang diabetes : {rate_women_diabetes:.2f}%\n")

    print(f"Persentase pria yang tidak diabetes : {rate_men_no_diabetes:.2f}%")
    print(f"Persentase pria yang prediabetes : {rate_men_prediabetes:.2f}%")
    print(f"Persentase pria yang diabetes : {rate_men_diabetes:.2f}%")
else:
    print("Error: Kolom 'Sex' atau 'Diabetes_012' tidak ditemukan dalam dataset.")
```

## **Mengklasifikasikan BMI**

```python
 # Import Library

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Import Data
df = pd.read_csv('/content/Data EDA.csv')
df.head()
```

```python
# Mengecek jumlah NaN pada kolom 'BMI'
nan_count = df['BMI'].isna().sum()
nan_count
```

```python
# Menampilkan total jumlah data
total_data = len(df)
print(f"Total jumlah data: {total_data}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Function to classify BMI
def classify_bmi(bmi):
    if pd.isna(bmi):  # Jika BMI adalah NaN, kembalikan kategori yang sesuai
        return 'Unknown'
    elif bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 24.9:
        return 'Normal weight'
    elif 25 <= bmi < 29.9:
        return 'Overweight'
    else:
        return 'Obesity'

# Cek dan hapus data NaN dalam kolom BMI
df = df.dropna(subset=['BMI'])  # Menghapus baris yang memiliki NaN pada kolom BMI

# Menambahkan kolom klasifikasi BMI
df['BMI Category'] = df['BMI'].apply(classify_bmi)

# Plot distribusi kategori BMI
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='BMI Category', palette='pastel')
plt.title("Distribution of BMI Categories")
plt.xlabel("BMI Category")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# Print jumlah kategori BMI dan pastikan totalnya sesuai
bmi_counts = df['BMI Category'].value_counts()
print("\nJumlah kategori BMI:")
print(bmi_counts)
print(f"\nTotal data (BMI): {len(df)}")  # Menampilkan total jumlah data yang valid
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Asumsi df sudah memiliki kolom 'BMI' dan 'Diabetes_012'

# Menambahkan kolom untuk klasifikasi BMI
def classify_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 24.9:
        return 'Normal weight'
    elif 25 <= bmi < 29.9:
        return 'Overweight'
    else:
        return 'Obesity'

# Menambahkan kolom klasifikasi BMI
df['BMI Category'] = df['BMI'].apply(classify_bmi)

# Menambahkan kolom untuk status diabetes
df['Diabetes_Status'] = df['Diabetes_012'].map({0: 'No Diabetes', 1: 'Prediabetes', 2: 'Diabetes'})

# Mengelompokkan berdasarkan kategori BMI dan status diabetes
bmi_diabetes_counts = df.groupby(['BMI Category', 'Diabetes_Status']).size().unstack(fill_value=0)

# Menampilkan hasil
print(bmi_diabetes_counts)

# Visualisasi distribusi BMI dan status diabetes menggunakan heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(bmi_diabetes_counts, annot=True, cmap="Blues", fmt="d", cbar=False)
plt.title("Distribusi Diabetes Berdasarkan Kategori BMI")
plt.xlabel("Diabetes Status")
plt.ylabel("BMI Category")
plt.tight_layout()
plt.show()
```

Berdasarkan tabel distribusi diatas, kita dapat melihat jumlah orang pada masing masing kategori klasifikasi BMI dan klasifikasi Diabetes. Dari tabel diatas pula, dapat dilihat jumlah orang terbanyak yaitu orang yang overweight namun tidak diabetes yaitu sebesar 81531 orang. Lalu orang yang terkena diabetes paling banyak ada pada orang yang obesitas dengan jumlah 20557.
```python
# Mengecek jumlah NaN pada kolom 'BMI'
 # Import Library

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Import Data
df = pd.read_csv('/content/Data EDA.csv')

nan_count = df['Age'].isna().sum()
nan_count
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Fungsi untuk mengklasifikasikan umur berdasarkan angka 1-13
def classify_age(age_code):
    mapping = {
        1: '18-24',
        2: '25-29',
        3: '30-34',
        4: '35-39',
        5: '40-44',
        6: '45-49',
        7: '50-54',
        8: '55-59',
        9: '60-64',
        10: '65-69',
        11: '70-74',
        12: '75-79',
        13: '80 or older'
    }
    return mapping.get(age_code)

# Menambahkan kolom untuk kategori umur
df['Age Category'] = df['Age'].apply(classify_age)

# Menghapus baris dengan nilai NaN pada kolom 'Age Category' (jika ada)
df = df.dropna(subset=['Age Category'])

# Menentukan urutan kategori umur
age_order = ['18-24', '25-29', '30-34', '35-39', '40-44', '45-49',
             '50-54', '55-59', '60-64', '65-69', '70-74', '75-79', '80 or older']

# Mengubah kolom 'Age Category' menjadi tipe kategori dengan urutan
df['Age Category'] = pd.Categorical(df['Age Category'], categories=age_order, ordered=True)

# Plot distribusi kategori umur
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Age Category', palette='Set2', order=age_order)
plt.title("Distribusi Kategori Umur")
plt.xlabel("Kategori Umur")
plt.ylabel("Jumlah Individu")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Print jumlah kategori umur
age_counts = df['Age Category'].value_counts()
print("\nJumlah kategori umur:")
print(age_counts)
```

Berdasarkan hasil dari barchart tersebut, ternyata kategori umur yang paling banyak diambil samplenya adalah pada rentang umur 65-59. Kategori umur yang paling sedikit diambil samplenya adalah pada rentang umur 18-24

# **Membuat Heatmap Korelasi**

```python
 # Import Library

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Import Data
df = pd.read_csv('/content/sample_data/Data EDA.csv')
df.head()
```

```python
# Pastikan data numerik dan abaikan NaN dalam perhitungan korelasi
Tcorrelation = df.apply(pd.to_numeric, errors='coerce').corr(method='pearson')
Tcorrelation
```

```python
# Pembuatan heatmap
plt.figure(figsize=(20, 16))
sns.heatmap(
    Tcorrelation,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    vmin=-1,
    vmax=1,
    xticklabels=Tcorrelation.columns,
    yticklabels=Tcorrelation.columns,
    linewidths=0.1,
    annot_kws={"size": 10} )
```

Berdasarkan heatmap korelasi tersebut maka diabetes disebabkan karena genetik dengan korelasi sebesar 0.31 dan sangat sedikit dipengaruhi oleh mental health, umur, dan jenis kelamin. Ia berbanding terbalik dengan konsumsi sayur, buah, tingkat aktivitas fisik, dan konsumsi alkohol.
